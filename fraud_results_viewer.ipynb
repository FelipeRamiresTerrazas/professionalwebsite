```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
import numpy as np

FIG_DIR = "figures"
figs = sorted(os.listdir(FIG_DIR))
print(f"Available figures in '{FIG_DIR}/':")
for f in figs:
    print(f"  • {f}")
```

## 1 — Exploratory Data Analysis (EDA)

```python
def show(filename, title=None):
    path = os.path.join(FIG_DIR, filename)
    if not os.path.exists(path):
        print(f"  Warning: file not found: {path}")
        return
    img = mpimg.imread(path)
    fig, ax = plt.subplots(figsize=(14, 8))
    ax.imshow(img)
    ax.axis('off')
    if title:
        ax.set_title(title, fontsize=14, fontweight='bold', pad=10)
    plt.tight_layout()
    plt.show()

show("01_class_imbalance.png", "Extreme Class Imbalance")
```

## 5 — Full Grid (All Charts)

```python
all_figs = sorted([f for f in os.listdir(FIG_DIR) if f.endswith('.png')])
n = len(all_figs)
cols = 2
rows = (n + 1) // cols

fig, axes = plt.subplots(rows, cols, figsize=(20, rows * 7))
axes = axes.flatten()

for i, fname in enumerate(all_figs):
    img = mpimg.imread(os.path.join(FIG_DIR, fname))
    axes[i].imshow(img)
    axes[i].axis('off')
    axes[i].set_title(fname.replace('.png', '').replace('_', ' ').title(),
                      fontsize=11, fontweight='bold')

# Hide empty axes
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

fig.suptitle('Credit Card Fraud Detection — All Results',
             fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```
